```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/creditcard.csv")
df.head()
```

***Check class distribution***

```python
plt.figure(figsize=(10, 5))
plt.title("Class Distribution")
sn.countplot(x="Class", data=df, palette="viridis", hue="Class", dodge=False, legend=False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xticks([0, 1], ["Non-Fraud (0)", "Fraud (1)"])
plt.show()
```

```python
# Data description
df.describe()
```

***Data processing***
- Cleaning the data
- Normalizing the columns Amount and Time

```python
print("Data shape: ", df.shape)
# Drop duplicates
df = df.drop_duplicates()
print("Data shape after drop duplicates: ", df.shape)
```

```python
# Normalization of amount and time variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Reshape the series to 2D array
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))

df.head()
```

Since fraud cases are rare, will:

- Undersample the majority class using NearMiss
- Oversample the minority class using SMOTE

> ðŸ‘‰ðŸ¾ Split data into features ***X*** and Target ***y***

```python
X = df.drop(columns=['Class'])
y = df['Class']
```

> ðŸ‘‰ðŸ¾ Use NearMiss to find the most frequent values in the variable `Class` and then use SMOTE to oversample the minority class.

```python
from imblearn.under_sampling import NearMiss
from collections import Counter
nearMiss = NearMiss()
X_resampled, y_resampled = nearMiss.fit_resample(X, y)
print("Class distribution after NearMiss:", Counter(y_resampled))
```

> ðŸ‘‰ðŸ¾ Use SMOTE oversampling

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(X, y)
print("Class distribution after SMOTE:", Counter(y_resampled))
```

***ðŸ‹ðŸ¾â€â™‚ï¸ Model Training***

Will train multiple classifiers:
- Logistic Regression
- Random Forest Classifier
- Support Vector Machine (SVM)
- Neural Network (MLP)

> ðŸ‘‰ðŸ¾ Split data for training and testing

```python
from sklearn.model_selection import train_test_split
# Split the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size=0.2, random_state=42, stratify=y_smote)
print("Training data shape: ", X_train.shape)
print("Testing data shape: ", X_test.shape)
```

***Train Logistic Regression***

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

lr = LogisticRegression()
lr.fit(X_train, y_train)

# Predict on the test set
lr_y_pred = lr.predict(X_test)

# Evaluate
print(classification_report(y_test, lr_y_pred))
```

***Train Random Forest***

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict on the test set
rf_y_pred = rf.predict(X_test)

# Evaluate the model
print(classification_report(y_test, rf_y_pred))
```

***Train Support Vector Machines***

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear')
svc.fit(X_train, y_train)

# Predict
svc_y_pred = svc.predict(X_test)

# Evaluate
print(classification_report(y_test, svc_y_pred))
```
